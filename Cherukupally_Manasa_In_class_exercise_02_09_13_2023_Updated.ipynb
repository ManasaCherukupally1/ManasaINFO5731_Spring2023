Question 2 (10 points): Write python code to collect 1000 data samples you discussed above.

```python
# You code here (Please add comments in the code):

import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the URL template for IMDb
base_url = "https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating"
pg_num = 1 #first page
tol_mov = 0

# Create empty lists
movie_names = []
movie_ratings = []
movie_genres = []

while tol_mov < 1000:
    # Create the full URL for the current page
    url = f"{base_url}&start={100 * (pg_num - 1)}"

    # Send a GET request
    response = requests.get(url)

    # Check for request
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the elements
        mov_ele = soup.find_all("div", class_="lister-item-content")

        # Iterate through the elements
        for movie_element in mov_ele:
            title = movie_element.h3.a.text
            rating = movie_element.strong.text
            genres = [genre.text.strip() for genre in movie_element.find_all("span", class_="genre")]

            movie_names.append(title)
            movie_ratings.append(rating)
            movie_genres.append(', '.join(genres))

            tol_mov += 1

        # Move to next page
        pg_num += 1

    else:
        print(f"Error: Unable to fetch data IMDb {pg_num}.")
        break

# Create DataFrame
data = {
    'Movie Name': movie_names,
    'Rating': movie_ratings,
    'Genres': movie_genres
}

df = pd.DataFrame(data)

# Print DataFrame
print(df)

```
